```python
## import all necessary packages and functions.
import csv # read and write csv files
from datetime import datetime # operations to parse dates
from pprint import pprint # use to print data structures like dictionaries in
                          # a nicer way than the base print function.
```

```python
def print_first_point(filename):
    """
    This function prints and returns the first data point (second row) from
    a csv file that includes a header row.
    """
    # print city name for reference
    city = filename.split('-')[0].split('/')[-1]
    print('\nCity: {}'.format(city))
    
    with open(filename, 'r') as f_in:
       
        ## TODO: Use the csv library to set up a DictReader object. ##
        ## see https://docs.python.org/3/library/csv.html           ##
        trip_reader = csv.DictReader(f_in)
        
        ## TODO: Use a function on the DictReader object to read the     ##
        ## first trip from the data file and store it in a variable.     ##
        ## see https://docs.python.org/3/library/csv.html#reader-objects #
        
        first_trip = next(trip_reader)
        
        ## TODO: Use the pprint library to print the first trip. ##
        ## see https://docs.python.org/3/library/pprint.html     ##
        
        pprint(first_trip)
        
    # output city name and first trip for later testing
    return (city, first_trip)

# list of files for each city
data_files = ['./data/NYC-CitiBike-2016.csv',
             './data/Chicago-Divvy-2016.csv',
              './data/Washington-CapitalBikeshare-2016.csv',]

# print the first trip from each file, store in dictionary
example_trips = {}
for data_file in data_files:
    city, first_trip = print_first_point(data_file)
    example_trips[city] = first_trip
    
```

### Condensing the Trip Data

It should also be observable from the above printout that each city provides different information. Even where the information is the same, the column names and formats are sometimes different. To make things as simple as possible when we get to the actual exploration, we should trim and clean the data. Cleaning the data makes sure that the data formats across the cities are consistent, while trimming focuses only on the parts of the data we are most interested in to make the exploration easier to work with.

You will generate new data files with five values of interest for each trip: trip duration, starting month, starting hour, day of the week, and user type. Each of these may require additional wrangling depending on the city:

- **Duration**: This has been given to us in seconds (New York, Chicago) or milliseconds (Washington). A more natural unit of analysis will be if all the trip durations are given in terms of minutes.
- **Month**, **Hour**, **Day of Week**: Ridership volume is likely to change based on the season, time of day, and whether it is a weekday or weekend. Use the start time of the trip to obtain these values. The New York City data includes the seconds in their timestamps, while Washington and Chicago do not. The [`datetime`](https://docs.python.org/3/library/datetime.html) package will be very useful here to make the needed conversions.
- **User Type**: It is possible that users who are subscribed to a bike-share system will have different patterns of use compared to users who only have temporary passes. Washington divides its users into two types: 'Registered' for users with annual, monthly, and other longer-term subscriptions, and 'Casual', for users with 24-hour, 3-day, and other short-term passes. The New York and Chicago data uses 'Subscriber' and 'Customer' for these groups, respectively. For consistency, you will convert the Washington labels to match the other two.

```python
def duration_in_mins(datum, city):
    
    """
    Takes as input a dictionary containing info about a single trip (datum) and
    its origin city (city) and returns the trip duration in units of minutes.
    
    Remember that Washington is in terms of milliseconds while Chicago and NYC
    are in terms of seconds. 
    
    HINT: The csv module reads in all of the data as strings, including numeric
    values. You will need a function to convert the strings into an appropriate
    numeric type when making your transformations.
    see https://docs.python.org/3/library/functions.html
    """
    if city == 'NYC':
        duration = int(datum['tripduration'])/60
    elif city == 'Chicago':
        duration = int(datum['tripduration'])/60
    else:
        duration = int(datum['Duration (ms)'])/60000
    return duration


# Some tests to check that your code works. There should be no output if all of
# the assertions pass. The `example_trips` dictionary was obtained from when
# you printed the first trip from each of the original data files.
tests = {'NYC': 13.9833,
         'Chicago': 15.4333,
         'Washington': 7.1231}

for city in tests:
    assert abs(duration_in_mins(example_trips[city], city) - tests[city]) < .001
```

```python
def time_of_trip(datum, city):
    """
    Takes as input a dictionary containing info about a single trip (datum) and
    its origin city (city) and returns the month, hour, and day of the week in
    which the trip was made.
    
    Remember that NYC includes seconds, while Washington and Chicago do not.
    
    HINT: You should use the datetime module to parse the original date
    strings into a format that is useful for extracting the desired information.
    see https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
    """
    
    if city == "NYC":
        time_object = datetime.strptime(datum['starttime'],'%m/%d/%Y %H:%M:%S')
        month = int(time_object.strftime('%m'))
        hour = int(time_object.strftime('%H')) 
        day_of_week = time_object.strftime('%A')    
        
    elif city == "Chicago":
        time_object = datetime.strptime(datum['starttime'],'%m/%d/%Y %H:%M')
        month = int(time_object.strftime('%m'))
        hour = int(time_object.strftime('%H'))
        day_of_week = time_object.strftime('%A')
        
    else:
        time_object = datetime.strptime(datum['Start date'],'%m/%d/%Y %H:%M')
        month = int(time_object.strftime('%m'))
        hour = int(time_object.strftime('%H'))
        day_of_week = time_object.strftime('%A')
    
    return (month, hour, day_of_week)


# Some tests to check that your code works. There should be no output if all of
# the assertions pass. The `example_trips` dictionary was obtained from when
# you printed the first trip from each of the original data files.
tests = {'NYC': (1, 0, 'Friday'),
        'Chicago': (3, 23, 'Thursday'),
         'Washington': (3, 22, 'Thursday')}

for city in tests:
    assert time_of_trip(example_trips[city], city) == tests[city]
    
```

```python
def type_of_user(datum, city):
    """
    Takes as input a dictionary containing info about a single trip (datum) and
    its origin city (city) and returns the type of system user that made the
    trip.
    
    Remember that Washington has different category names compared to Chicago
    and NYC. 
    """
    if city == 'NYC':
        user_type = datum['usertype']
    elif city == "Chicago":
        user_type = datum['usertype']
    else:
        w_user_type = datum['Member Type']
        if w_user_type == 'Registered':
            user_type = 'Subscriber'
        else:
            user_type = 'Customer'
    return user_type


# Some tests to check that your code works. There should be no output if all of
# the assertions pass. The `example_trips` dictionary was obtained from when
# you printed the first trip from each of the original data files.
tests = {'NYC': 'Customer',
         'Chicago': 'Subscriber',
         'Washington': 'Subscriber'}

for city in tests:
    assert type_of_user(example_trips[city], city) == tests[city]
```

```python
def number_of_trips(filename):
    """
    This function reads in a file with trip data and reports the number of
    trips made by subscribers, customers, and total overall along with proportion of 
    trips of subscribers and customers in total trips. 
    """
    with open(filename, 'r') as f_in:
        # set up csv reader object
        reader = csv.DictReader(f_in)
        
        # initialize count variables
        n_subscribers = 0
        n_customers = 0
        
        # tally up ride types
        for row in reader:
            if row['user_type'] == 'Subscriber':
                n_subscribers += 1
            else:
                n_customers += 1 
        # compute total number of rides
        n_total = n_subscribers + n_customers
        sub_prop = n_subscribers*100/n_total
        cust_prop = n_customers*100/n_total 
        
        # return tallies as a tuple
        return(n_total, n_subscribers, n_customers, sub_prop, cust_prop)
```

```python
## Use this and additional cells to answer Question 4b.                 ##
##                                                                      ##
## HINT: The csv module reads in all of the data as strings, including  ##
## numeric values. You will need a function to convert the strings      ##
## into an appropriate numeric type before you aggregate data.          ##
## TIP: For the Bay Area example, the average trip length is 14 minutes ##
## and 3.5% of trips are longer than 30 minutes.                        ##
#Total duration
#No. of trips
#Trips with duration >=30m
#prop of 30m rides = above value/no.of trips 

def avg_trip_length(file_name):
    
    """
    This function reads in a file with trip data and reports the total duration of trips
    by subscribers & by customers, average trip length and proporation of trips with duration 
    more than or equal to 30min in total no.of trips
    
    """
    
    with open (file_name,'r') as avg_file:
     
        trip_object = csv.DictReader(avg_file)
        no_of_trips = 0
        no_of_30m_trips = 0
        total_trip_duration = 0
        tot_sub_trip_duration = 0 
        tot_cust_trip_duration = 0 
    
        ## Calculating total trip duration for both subscribers & Customers
        ## Calculating total trip duration separately for both subscribers & Customers
        ## Calculating total no.of trips for both subscribers & Customers
        ## Calculating total no.of trips with duration >= 30min 
            
        for avg_trip in trip_object:
            trip_duration = float(avg_trip['duration'])  
            if avg_trip['user_type'] == 'Subscriber':
                tot_sub_trip_duration += trip_duration
            else:
                tot_cust_trip_duration += trip_duration 
            total_trip_duration += trip_duration    
            no_of_trips += 1  
            if trip_duration >= 30:
                no_of_30m_trips += 1
        
        ## Calculating average trip length       
        avg_length = total_trip_duration/no_of_trips
            
        ## Calculating proportion of 30min trips in total no.of trips 
        prop_30m_trips = no_of_30m_trips*100/no_of_trips
            
        return(tot_sub_trip_duration, tot_cust_trip_duration, avg_length, prop_30m_trips)
        
```

```python
trip_files = ['./data/Chicago-2016-Summary.csv', 
             './data/NYC-2016-Summary.csv',
             './data/Washington-2016-Summary.csv']
for trip_file in trip_files:
    avg_trip_dic= {}
    avg_trip_dic[trip_file.split('-')[0].split('/')[-1]] = avg_trip_length(trip_file)
    
    print("Average trip length in {} is {}".format(trip_file.split('-')[0].split('/')[-1],
                                                avg_trip_dic[trip_file.split('-')[0].split('/')[-1]][2]))
    print("{} proportion of rides are longer than 30min in {}".format(avg_trip_dic[trip_file.split('-')[0].split('/')[-1]][3],
                                                                trip_file.split('-')[0].split('/')[-1]))
```

```python
## Use this and additional cells to continue to explore the dataset. ##
## Once you have performed your exploration, document your findings  ##
## in the Markdown cell above.    
## Is the pattern of ridership different on the weekends versus weekdays? 
## On what days are Subscribers most likely to use the system? 
## What about Customers? Does the average duration of rides change depending on the day of the week?

def trip_visual_by_dow(file_name):
    
    """
    This funtion taking the input of file with trip data returns dictionaries of total trips,
    subscriber trips, customer trips and average trip durations which are used to 
    show visualizations 
   
    """
    #Mon no.of trips & duration for customers & Subscribers 
    mon_sub_trips = 0 
    mon_cust_trips = 0 
    mon_sub_trip_dur = 0 
    mon_cust_trip_dur = 0
    #Tue no.of trips & duration for customers & Subscribers 
    tue_sub_trips = 0
    tue_cust_trips = 0
    tue_sub_trip_dur = 0
    tue_cust_trip_dur = 0
    #Wed no.of trips & duration for customers & Subscribers 
    wed_sub_trips = 0
    wed_cust_trips = 0
    wed_sub_trip_dur = 0
    wed_cust_trip_dur = 0
    #Thu no.of trips & duration for customers & Subscribers 
    thu_sub_trips = 0
    thu_cust_trips = 0
    thu_sub_trip_dur =0 
    thu_cust_trip_dur = 0
    #Fri no.of trips & duration for customers & Subscribers 
    fri_sub_trips = 0
    fri_cust_trips = 0
    fri_sub_trip_dur = 0
    fri_cust_trip_dur = 0
    #Sat no.of trips & duration for customers & Subscribers 
    sat_sub_trips = 0
    sat_cust_trips = 0
    sat_sub_trip_dur= 0
    sat_cust_trip_dur = 0
    #Sun no.of trips & duration for customers & Subscribers 
    sun_sub_trips = 0
    sun_cust_trips = 0
    sun_sub_trip_dur = 0
    sun_cust_trip_dur = 0
    

    
    with open (file_name, 'r') as f_in:
        dow_visual = csv.DictReader(f_in)  
            
        for dow in dow_visual:
            if dow['day_of_week'] == 'Monday':
                if dow['user_type'] =='Subscriber':
                    mon_sub_trips += 1
                    mon_sub_trip_dur += float(dow['duration'])
                else:
                    mon_cust_trips += 1
                    mon_cust_trip_dur += float(dow['duration']) 
            elif dow['day_of_week'] == 'Tuesday':
                if dow['user_type'] =='Subscriber':
                    tue_sub_trips += 1
                    tue_sub_trip_dur += float(dow['duration'])
                else:
                    tue_cust_trips += 1
                    tue_cust_trip_dur += float(dow['duration'])
            elif dow['day_of_week'] == 'Wednesday':
                if dow['user_type'] =='Subscriber':
                    wed_sub_trips += 1
                    wed_sub_trip_dur += float(dow['duration'])
                else:
                    wed_cust_trips += 1
                    wed_cust_trip_dur += float(dow['duration'])
            elif dow['day_of_week'] == 'Thursday':
                if dow['user_type'] =='Subscriber':
                    thu_sub_trips += 1
                    thu_sub_trip_dur += float(dow['duration'])
                else:
                    thu_cust_trips += 1
                    thu_cust_trip_dur += float(dow['duration'])
            elif dow['day_of_week'] == 'Friday':
                if dow['user_type'] =='Subscriber':
                    fri_sub_trips += 1
                    fri_sub_trip_dur += float(dow['duration'])
                else:
                    fri_cust_trips += 1
                    fri_cust_trip_dur += float(dow['duration'])
            elif dow['day_of_week'] == 'Saturday':
                if dow['user_type'] =='Subscriber':
                    sat_sub_trips += 1
                    sat_sub_trip_dur += float(dow['duration'])
                else:
                    sat_cust_trips += 1
                    sat_cust_trip_dur += float(dow['duration'])
            else: 
                dow['day_of_week'] == 'Sunday'
                if dow['user_type'] =='Subscriber':
                    sun_sub_trips += 1
                    sun_sub_trip_dur += float(dow['duration'])
                else:
                    sun_cust_trips += 1
                    sun_cust_trip_dur += float(dow['duration'])
                    
            
        #Monday avg trip duration calculation group wise and cumulatively 
        mon_tot_trip_dur = mon_sub_trip_dur + mon_cust_trip_dur
        mon_tot_trips = mon_sub_trips + mon_cust_trips
        mon_avg_tot_trip_dur = mon_tot_trip_dur/mon_tot_trips
        mon_avg_sub_trip_dur =  mon_sub_trip_dur/mon_sub_trips
        mon_avg_cust_trip_dur = mon_cust_trip_dur/mon_cust_trips
            
        #Tuesday avg trip duration calculation group wise and cumulatively
        tue_tot_trip_dur = tue_sub_trip_dur + tue_cust_trip_dur
        tue_tot_trips = tue_sub_trips + tue_cust_trips
        tue_avg_tot_trip_dur = tue_tot_trip_dur/tue_tot_trips
        tue_avg_sub_trip_dur =  tue_sub_trip_dur/tue_sub_trips
        tue_avg_cust_trip_dur = tue_cust_trip_dur/tue_cust_trips
            
        #Wednesday avg trip duration calculation group wise and cumulatively
        wed_tot_trip_dur = wed_sub_trip_dur + wed_cust_trip_dur
        wed_tot_trips = wed_sub_trips + wed_cust_trips
        wed_avg_tot_trip_dur = wed_tot_trip_dur/wed_tot_trips
        wed_avg_sub_trip_dur =  wed_sub_trip_dur/wed_sub_trips
        wed_avg_cust_trip_dur = wed_cust_trip_dur/wed_cust_trips
            
        #Thursday avg trip duration calculation group wise and cumulatively
        thu_tot_trip_dur = thu_sub_trip_dur + thu_cust_trip_dur
        thu_tot_trips = thu_sub_trips + thu_cust_trips
        thu_avg_tot_trip_dur = thu_tot_trip_dur/thu_tot_trips
        thu_avg_sub_trip_dur =  thu_sub_trip_dur/thu_sub_trips
        thu_avg_cust_trip_dur = thu_cust_trip_dur/thu_cust_trips
            
        #Friday avg trip duration calculation group wise and cumulatively
        fri_tot_trip_dur = fri_sub_trip_dur + fri_cust_trip_dur
        fri_tot_trips = fri_sub_trips + fri_cust_trips
        fri_avg_tot_trip_dur = fri_tot_trip_dur/fri_tot_trips
        fri_avg_sub_trip_dur =  fri_sub_trip_dur/fri_sub_trips
        fri_avg_cust_trip_dur = fri_cust_trip_dur/fri_cust_trips
            
        #Saturday avg trip duration calculation group wise and cumulatively
        sat_tot_trip_dur = sat_sub_trip_dur + sat_cust_trip_dur
        sat_tot_trips = sat_sub_trips + sat_cust_trips
        sat_avg_tot_trip_dur = sat_tot_trip_dur/sat_tot_trips
        sat_avg_sub_trip_dur =  sat_sub_trip_dur/sat_sub_trips
        sat_avg_cust_trip_dur = sat_cust_trip_dur/sat_cust_trips
            
        #Sunday avg trip duration calculation group wise and cumulatively
        sun_tot_trip_dur = sun_sub_trip_dur + sun_cust_trip_dur
        sun_tot_trips = sun_sub_trips + sun_cust_trips
        sun_avg_tot_trip_dur = sun_tot_trip_dur/sun_tot_trips
        sun_avg_sub_trip_dur =  sun_sub_trip_dur/sun_sub_trips
        sun_avg_cust_trip_dur = sun_cust_trip_dur/sun_cust_trips
            
        ridership_dic = {"Mon": mon_tot_trips,"Tue":tue_tot_trips,"Wed": wed_tot_trips,"Thu":thu_tot_trips,
                            "Fri":fri_tot_trips,"Sat":sat_tot_trips,"Sun":sun_tot_trips}
            
        subscriber_dic = {"Mon": mon_sub_trips,"Tue":tue_sub_trips,"Wed": wed_sub_trips,"Thu":thu_sub_trips,
                            "Fri":fri_sub_trips,"Sat":sat_sub_trips,"Sun":sun_sub_trips}

        customer_dic = {"Mon": mon_cust_trips,"Tue":tue_cust_trips,"Wed": wed_cust_trips,"Thu":thu_cust_trips,
                            "Fri":fri_cust_trips,"Sat":sat_cust_trips,"Sun":sun_cust_trips}
            
        avg_duration_dic = {"Mon": mon_avg_tot_trip_dur,"Tue":tue_avg_tot_trip_dur,"Wed": wed_avg_tot_trip_dur,
                                "Thu":thu_avg_tot_trip_dur,"Fri":fri_avg_tot_trip_dur,"Sat":sat_avg_tot_trip_dur,
                                "Sun":sun_avg_tot_trip_dur}
            
        return(ridership_dic,subscriber_dic,customer_dic,avg_duration_dic)
        

file = "./data/NYC-2016-Summary.csv"

ridership_visual = trip_visual_by_dow(file)[0]
subscriber_visual = trip_visual_by_dow(file)[1]
customer_visual = trip_visual_by_dow(file)[2]
avg_duration_visual = trip_visual_by_dow(file)[3]

day_of_week = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]

x =  [ k for k in customer_visual ]
y = [v for v in customer_visual.values()]
x_pos =np.arange(len(x))

x2 = [ k for k in subscriber_visual]
y2 = [v for v in subscriber_visual.values()]
x2_pos =np.arange(len(x2))

plt.bar(x,y,  alpha=1.0, label ='Subscriber', color = 'r')
plt.bar(x2,y2, alpha=0.5, label ='Customer', color ='g')

plt.title('Ridership frequency of Subscribers vs Customers in NYC')
plt.xlabel('Day of the Week')
plt.ylabel('Ridership Frequency')
plt.legend()
plt.show()


x3 = [ k for k in avg_duration_visual]
y3 = [v for v in avg_duration_visual.values()]

plt.bar(x3,y3, alpha=0.5, color ='y')
plt.title('Average duration of trip for each of day of week in NYC')
plt.xlabel('Day of the Week')
plt.ylabel('Average duration')
plt.show()


x4 = [k for k in ridership_visual]
y4 = [v for v in ridership_visual.values()]

plt.bar(x4,y4, alpha=0.5, color ='b')
plt.title('No.of trips for each of day week in NYC')
plt.xlabel('Day of the Week')
plt.ylabel('Number of trips')
plt.show()

        
    
```
